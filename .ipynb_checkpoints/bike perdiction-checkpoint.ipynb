```python
def cleaner(num):
    try:
        i = re.sub('[^0-9]',"",num)
        return i
    except:
        pass
```

```python
bike['km_driven'] = bike['km_driven'].apply(lambda x:cleaner(x))
bike['bike_price'] = bike['bike_price'].apply(lambda x : cleaner(x))
```

```python
bike['km_driven'].unique()
```

```python
bike['bike_price'].unique()
```

```python
bike['bike_price'] = bike['bike_price'].astype('float64')
bike['km_driven'] = bike['km_driven'].astype('float64')
```

```python
bike
```

```python
bike['bike_address'] = bike['bike_address'].str.split(",").str.get(1)
bike
```

```python
bike.isna().sum()
```

```python
bike = bike.dropna(subset="bike_price")
bike
```

```python
bike.isna().sum()
```

```python
cat_data = bike.select_dtypes(include="object")
cat_data
```

```python
for i in cat_data.columns:
    cat_data[i] = cat_data[i].str.lower()
```

```python
cat_data
```

```python
num_data = bike.select_dtypes(include=np.number)
print("The numerical data:-",num_data.columns)
print("The categorical data:-",cat_data.columns)
```

```python
num_data.isnull().sum()
```

```python
cat_data.isnull().sum()
```

```python
for i in num_data.columns:
    sns.boxplot(num_data[i])
    plt.show()
```

```python
for i in num_data.columns:
    x=[num_data[i].isnull()]
    if x==0:
        pass
    else:
        num_data[i].fillna(num_data[i].median(),inplace=True)
```

```python
for i in cat_data.columns:
    x=[cat_data[i].isnull()]
    if x==0:
        pass
    else:
        cat_data[i].fillna(cat_data[i].mode()[0],inplace=True)
```

```python
num_data.isnull().sum()
```

```python
cat_data.isnull().sum()
```

```python
bike = pd.concat([num_data,cat_data],1)
bike
```

```python
bike.rename(columns={'bike_price':'price'},inplace=True)
bike
```

```python
for i in bike.columns:
    try:
        print('The column name:-',i)
        sns.boxplot(bike[i],color='lightgreen')
        plt.show()
    except:
        pass
```

```python
from feature_engine.outliers import Winsorizer
km = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='km_driven')
bike['km_driven'] = km.fit_transform(bike[['km_driven']])
year = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='make_year')
bike['make_year'] = year.fit_transform(bike[['make_year']])
```

```python
for i in bike.columns:
    try:
        print('The column name:-',i)
        sns.boxplot(bike[i],color='lightgreen')
        plt.show()
    except:
        pass
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
a = ohe.fit_transform(bike[['bike_name']])
b = ohe.get_feature_names_out()
c = ohe.fit_transform(bike[['bike_model']])
d = ohe.get_feature_names_out()
e = ohe.fit_transform(bike[['bike_address']])
f = ohe.get_feature_names_out()
g = ohe.fit_transform(bike[['state']])
h = ohe.get_feature_names_out()
x1 = pd.DataFrame(a,columns=b)
x2 = pd.DataFrame(c,columns=d)
x3 = pd.DataFrame(e,columns=f)
x4 = pd.DataFrame(g,columns=h)
```

```python
x1.shape,x2.shape,x3.shape,x4.shape
```

```python
final = pd.concat([x1,x2,x3,x4],axis=1)
final
```

```python
final.reset_index(inplace=True)
bike.reset_index(inplace=True)
```

```python
bike1 = pd.concat([bike,final],axis=1)
bike1
```

```python
bike1.drop(columns = {'bike_name','bike_model','bike_address','state','index'},inplace=True)
bike1
```

```python
bike1.info()
```

```python
x = bike1.drop('price',1)
y = bike1['price']
x.shape,y.shape
```

```python
bike1.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
model =  RandomForestRegressor()
model.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model.score(x_train,y_train)
```
