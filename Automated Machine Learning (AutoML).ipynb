```python
def cleaner(num):
    try:
        i = re.sub('[^0-9]',"",num)
        return i
    except:
        pass
```

```python
bike['km_driven'] = bike['km_driven'].apply(lambda x:cleaner(x))
bike['bike_price'] = bike['bike_price'].apply(lambda x : cleaner(x))
```

```python
best_pipeline=automl.best_pipeline
```
